```python
# As you can see, variables are stored and can be run in different blocks
# you can also import packages and install new ones (see the import below)

# Function definitons can also be saved and run later. The function
# below creates a csv file starting from a dictionary in which the keys
# are the colums of the csv, and the values are the rows.

# example b = {"first":[1,2,3], "second":[2,3,5]}
# will create a csv that looks like this
#   first,second
#     1,2
#     2,3
#     3,5

import csv

def csvfromdict(dictionary):
  head = dictionary.keys()
  maxlen = 0
  for key in dictionary:
    if len(dictionary[key]) > maxlen:
      maxlen = len(dictionary[key])
  idx = 0
  with open("result.csv", "w+") as output:
    writer = csv.writer(output)
    writer.writerow(head)
    while idx < maxlen:
      row = []
      for key in dictionary:
        if idx < len(dictionary[key]):
          row.append(dictionary[key][idx])
        else:
          row.append("")
      writer.writerow(row)
      idx +=1
  print("done")
    
```

```python
# Now run this code, it will create a dictionary and call the csvfromdict 
# function.

b = {"first":[1,2,8], "second":[2,3,5]}

csvfromdict(b)
```
